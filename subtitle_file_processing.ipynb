```python
from datetime import datetime, timedelta
import tqdm
```

```python
filepath = '/home/archeron/Videos/Good Bye Lenin (2003) 720p_WEB_DL_sujaidr/Good.Bye.Lenin.2003.German.1080p.BluRay.x264-DETAiLS.srt'

srt = open(filepath, 'r').read()
```

```python
# '1\n00:00:47,567 --> 00:00:51,401\n(Vater)\nGuck mal her. Hier in die Kamera.\n\n'
# => '1\n00:00:32,567 --> 00:00:36,401\n(Vater)\nGuck mal her. Hier in die Kamera.\n\n
def replace_line(line):
    pieces = line.split('\n')
    replaced_line =  '\n'.join([pieces[0], replace_time(pieces[1]), *pieces[2:]])
    return replaced_line

# 00:00:47,567 --> 00:00:51,401
# => 00:00:32,567 --> 00:00:36,401
def replace_time(total_timestring):
    time_list = []
    for timestring in total_timestring.split(' --> '):
        # convert to datetime object 
        time_obj = datetime.strptime(timestring.split(',')[0], '%H:%M:%S')
        # subtracting 15secs using the time delta function
        new_time = time_obj - timedelta(seconds=15)
        # convert back to string and join with the old 
        new_time_string_joined = ','.join([str(new_time.time()), timestring.split(',')[1]])
        time_list.append(new_time_string_joined)
    return ' --> '.join(time_list)
```

```python
lines_list = []
for line in tqdm.notebook.tqdm(srt.split('\n\n')):
    try:
        new_line = replace_line(line)
        lines_list.append(new_line)
    except:
        pass
new_srt = '\n\n'.join(lines_list)
```

```python
write_filepath = '/home/archeron/Videos/Good Bye Lenin (2003) 720p_WEB_DL_sujaidr/Good.Bye.Lenin.2003.German.1080p.BluRay.x264-DETAiLS_processed.srt'

write_file = open(write_filepath,"w").write(new_srt)
```
